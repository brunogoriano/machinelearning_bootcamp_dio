```python
from google.colab.patches import cv2_imshow
import cv2
```

```python
# Carregar a imagem colorida
imagem_colorida = cv2.imread('cachorro.jpg')
```

```python
# Converter a imagem para níveis de cinza
imagem_cinza = cv2.cvtColor(imagem_colorida, cv2.COLOR_BGR2GRAY)
```

```python
# Binarizar a imagem (definir um limiar)
limiar = 127  # Valor de limiar (ajuste conforme necessário)
_, imagem_binaria = cv2.threshold(imagem_cinza, limiar, 255, cv2.THRESH_BINARY)
```

```python
# Exibir as imagens
print("Imagem Colorida:")
cv2_imshow(imagem_colorida)
```

```python
# Salvar as imagens resultantes (opcional)
cv2.imwrite('imagem_cinza.jpg', imagem_cinza)
cv2.imwrite('imagem_binaria.jpg', imagem_binaria)
```

```python
print("Imagem em Níveis de Cinza:")
cv2_imshow(imagem_cinza)
```

```python
print("Imagem Binarizada:")
cv2_imshow(imagem_binaria)
```
